```python
import pandas as pd
import re

# ======================================================
# 1) νμΌ λ¶λ¬μ¤κΈ°
# ======================================================
file = "/content/μ§€μ—­λ³„_μ•μ „ λΉ„μƒλ²¨ μ •λ³΄.xlsx"
df = pd.read_excel(file)

# μ»¬λΌλ… μλ™ μ •λ¦¬(μ–‘μ½ κ³µλ°± μ κ±°)
df.columns = df.columns.str.strip()

# μ£Όμ† κ΄€λ ¨ μ»¬λΌλ… μ„¤μ •
col_road = "μ†μ¬μ§€λ„λ΅λ…μ£Όμ†"
col_addr = "μ†μ¬μ§€μ§€λ²μ£Όμ†"
col_manage = "κ΄€λ¦¬κΈ°κ΄€λ…"
col_year = "μ•μ „λΉ„μƒλ²¨μ„¤μΉμ—°λ„"

# ======================================================
# 2) μ„μΈ 25κ° κµ¬ λ¦¬μ¤νΈ
# ======================================================
seoul_gu_list = [
    "μΆ…λ΅κµ¬","μ¤‘κµ¬","μ©μ‚°κµ¬","μ„±λ™κµ¬","κ΄‘μ§„κµ¬","λ™λ€λ¬Έκµ¬","μ¤‘λ‘κµ¬","μ„±λ¶κµ¬","κ°•λ¶κµ¬","λ„λ΄‰κµ¬",
    "λ…Έμ›κµ¬","μ€ν‰κµ¬","μ„λ€λ¬Έκµ¬","λ§ν¬κµ¬","μ–‘μ²κµ¬","κ°•μ„κµ¬","κµ¬λ΅κµ¬","κΈμ²κµ¬","μλ“±ν¬κµ¬",
    "λ™μ‘κµ¬","κ΄€μ•…κµ¬","μ„μ΄κµ¬","κ°•λ‚¨κµ¬","μ†΅νκµ¬","κ°•λ™κµ¬"
]

# ======================================================
# 3) μ£Όμ†μ—μ„ κµ¬ μ¶”μ¶ ν•¨μ
# ======================================================
def extract_gu(text):
    if pd.isna(text):
        return None
    pattern = r"(μΆ…λ΅κµ¬|μ¤‘κµ¬|μ©μ‚°κµ¬|μ„±λ™κµ¬|κ΄‘μ§„κµ¬|λ™λ€λ¬Έκµ¬|μ¤‘λ‘κµ¬|μ„±λ¶κµ¬|κ°•λ¶κµ¬|λ„λ΄‰κµ¬|" \
              r"λ…Έμ›κµ¬|μ€ν‰κµ¬|μ„λ€λ¬Έκµ¬|λ§ν¬κµ¬|μ–‘μ²κµ¬|κ°•μ„κµ¬|κµ¬λ΅κµ¬|κΈμ²κµ¬|μλ“±ν¬κµ¬|" \
              r"λ™μ‘κµ¬|κ΄€μ•…κµ¬|μ„μ΄κµ¬|κ°•λ‚¨κµ¬|μ†΅νκµ¬|κ°•λ™κµ¬)"
    m = re.search(pattern, str(text))
    return m.group(1) if m else None

# ======================================================
# 4) μµμΆ… μ§€μ—­ μ¶”μ¶ ν•¨μ
# ======================================================
def get_final_region(row):
    r1 = extract_gu(row[col_road])
    if r1 in seoul_gu_list:
        return r1

    r2 = extract_gu(row[col_addr])
    if r2 in seoul_gu_list:
        return r2

    r3 = extract_gu(row[col_manage])
    if r3 in seoul_gu_list:
        return r3

    return "κΈ°νƒ€"

df["μµμΆ…μ§€μ—­"] = df.apply(get_final_region, axis=1)

# ======================================================
# 5) κΈ°νƒ€ λ©λ΅ λ° μ§€μ—­λ³„ κ°μ μ¶λ ¥
# ======================================================
κΈ°νƒ€κ°’ = df[df["μµμΆ…μ§€μ—­"]=="κΈ°νƒ€"]["μµμΆ…μ§€μ—­"].unique()
print("----- κΈ°νƒ€λ΅ λ¶„λ¥λ κ°’ λ©λ΅ -----")
print(κΈ°νƒ€κ°’)

print("\n----- μ§€μ—­λ³„ κ°μ -----")
print(df["μµμΆ…μ§€μ—­"].value_counts())

# ======================================================
# 6) μ£Όμ†/κ΄€λ¦¬κΈ°κ΄€λ… μ „μ²΄μ—μ„ λ“±μ¥ν• κµ¬λ¥Ό μ €μ¥ (μ΄μƒμΉ κ²€μ¶μ©)
# ======================================================
def find_all_gu(text):
    if pd.isna(text):
        return []
    pattern = r"(μΆ…λ΅κµ¬|μ¤‘κµ¬|μ©μ‚°κµ¬|μ„±λ™κµ¬|κ΄‘μ§„κµ¬|λ™λ€λ¬Έκµ¬|μ¤‘λ‘κµ¬|μ„±λ¶κµ¬|κ°•λ¶κµ¬|λ„λ΄‰κµ¬|" \
              r"λ…Έμ›κµ¬|μ€ν‰κµ¬|μ„λ€λ¬Έκµ¬|λ§ν¬κµ¬|μ–‘μ²κµ¬|κ°•μ„κµ¬|κµ¬λ΅κµ¬|κΈμ²κµ¬|μλ“±ν¬κµ¬|" \
              r"λ™μ‘κµ¬|κ΄€μ•…κµ¬|μ„μ΄κµ¬|κ°•λ‚¨κµ¬|μ†΅νκµ¬|κ°•λ™κµ¬)"
    return re.findall(pattern, str(text))

df["μ£Όμ†λ‚΄_λ¨λ“ _κµ¬"] = df.apply(
    lambda x: find_all_gu(f"{x[col_road]} {x[col_addr]} {x[col_manage]}"),
    axis=1
)

df_problem = df[df.apply(
    lambda x: x["μµμΆ…μ§€μ—­"] not in x["μ£Όμ†λ‚΄_λ¨λ“ _κµ¬"] and x["μµμΆ…μ§€μ—­"] != "κΈ°νƒ€",
    axis=1
)]

print("\n μ£Όμ†μ—μ„ μ¶”μ¶ν• κµ¬κ°€ μ‹¤μ  λ¬Έμμ—΄κ³Ό λ¶μΌμΉν• ν–‰ μ:", len(df_problem))
print(df_problem[[col_road, col_addr, col_manage, "μ£Όμ†λ‚΄_λ¨λ“ _κµ¬", "μµμΆ…μ§€μ—­"]].head(20))

# ======================================================
# 7) μ„¤μΉλ€μ κ³„μ‚° λ° 2025λ…„ μ΄ν›„ ν•©μ‚°
# ======================================================
df_clean = df.dropna(subset=[col_year]).copy()
df_clean["μ„¤μΉλ€μ"] = 1

# 2025λ…„ μ΄κ³Ό μ—°λ„ λ°μ΄ν„° μ¶”μ¶
over_2025 = df[df[col_year] > 2025]

# μ„¤μΉ μ—°λ„μ™€ μµμΆ…μ§€μ—­ ν™•μΈ
over_2025_info = over_2025[[col_year, "μµμΆ…μ§€μ—­", col_road, col_addr, col_manage]]

print("----- 2025λ…„ μ΄κ³Ό μ„¤μΉ μ—°λ„ λ°μ΄ν„° -----")
print(over_2025_info)

print("\n 2025λ…„ μ΄κ³Ό μ—°λ„ μ΅΄μ¬ν•λ” μ§€μ—­ λ©λ΅:")
print(over_2025_info["μµμΆ…μ§€μ—­"].value_counts())


# 2025λ…„ μ΄κ³Ό μ—°λ„λ” 2025λ΅ ν•©μΉκΈ°
df_clean[col_year] = df_clean[col_year].apply(lambda x: 2025 if x > 2025 else x)

# μ—°λ„λ³„ + μ§€μ—­λ³„ μ§‘κ³„
year_region = df_clean.groupby([col_year, "μµμΆ…μ§€μ—­"])["μ„¤μΉλ€μ"].sum().reset_index()
year_region = year_region.sort_values(by=[col_year, "μµμΆ…μ§€μ—­"])

# λ„μ  μ„¤μΉλ€μ κ³„μ‚°
year_region["λ„μ μ„¤μΉλ€μ"] = year_region.groupby("μµμΆ…μ§€μ—­")["μ„¤μΉλ€μ"].cumsum()

# ======================================================
# 9) CSV μƒμ„±
# ======================================================
year_region.to_csv("λΉ„μƒλ²¨_λ…„λ„λ³„_μ§€μ—­λ³„_λ„μ μ„¤μΉλ€μ_2025ν•©μ‚°.csv", index=False)
print("\n CSV μƒμ„± μ™„λ£: λΉ„μƒλ²¨_λ…„λ„λ³„_μ§€μ—­λ³„_λ„μ μ„¤μΉλ€μ_2025ν•©μ‚°.csv")
```
